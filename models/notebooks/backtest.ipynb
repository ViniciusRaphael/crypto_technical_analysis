```python

input_path = r'D:\Github\Forked\crypto_technical_analysis\files\crypto_data_prep_models.parquet'
dados_prep = pd.read_parquet(input_path)
dados_prep
```

```python
pd.set_option("display.max_columns", None)

dados_filter = dados_prep[['Symbol', 'Date', 'Volume', 'Close', 'target_10_7d', 'target_10_15d', 'target_10_30d']]
dados_score = dados_proba[['Symbol', 'Date', 'lr_pb_10_7d', 'lr_pb_15_7d', 'lr_pb_20_7d', 'lr_pb_25_7d', 'lr_pb_10_15d', 'lr_pb_15_15d', 'lr_pb_20_15d', 'lr_pb_25_15d', 'lr_pb_10_30d', 'lr_pb_15_30d', 'lr_pb_20_30d', 'lr_pb_25_30d', 'score']]
```

```python
import numpy as np

df_merged = pd.merge(dados_filter, dados_score, on=['Date', 'Symbol'], how='inner')
print(len(df_merged))
# Substituindo valores infinitos por NaN
df_merged.replace([np.inf, -np.inf], np.nan, inplace=True)
print(len(df_merged))

# Removendo linhas com valores NaN
df_merged.dropna(inplace=True)
print(len(df_merged))
df_merged.head(2)
```

```python
## Entradas
start_date = '2024-03-01'
# today_date = datetime.today().strftime('%Y-%m-%d')
last_date = str(df_merged['Date'].max())

# Gerar um range de datas
datas = pd.date_range(start=start_date, end=last_date, freq='15D')

# Converter para formato YYYY-MM-DD
datas_formatadas = datas.strftime('%Y-%m-%d')

output_dataset = pd.DataFrame()

for data in datas_formatadas:



    df_merged_score_clean_cumulative = df_merged[(df_merged['lr_pb_10_15d'] >= 0.7) & (df_merged['Symbol'] != 'NOME-USD')  & (df_merged['Date'] == data)]

    df_merged_score_clean_cumulative = df_merged_score_clean_cumulative.sort_values(by='lr_pb_10_15d', ascending=False).head(2)

    output_dataset = pd.concat([output_dataset, df_merged_score_clean_cumulative])

output_dataset['Cumulative_Return_7d'] = (1 + ( output_dataset['target_10_7d'])).cumprod() - 1
output_dataset['Cumulative_Return_15d'] = (1 + ( output_dataset['target_10_15d'])).cumprod() - 1
output_dataset['Cumulative_Return_30d'] = (1 + ( output_dataset['target_10_30d'])).cumprod() - 1


output_dataset
```

```python
import os

# Caminho para a pasta contendo os arquivos CSV
caminho_pasta = 'results/'

# Listar todos os arquivos CSV na pasta
arquivos_csv = [f for f in os.listdir(caminho_pasta) if f.endswith('.csv')]
```

```python
arquivos_csv
```

```python
# Inicializar uma lista para armazenar os DataFrames
dataframes = []

# Ler cada arquivo CSV e adicioná-lo à lista de DataFrames
for arquivo in arquivos_csv:
    caminho_arquivo = os.path.join(caminho_pasta, arquivo)
    df = pd.read_csv(caminho_arquivo)
    dataframes.append(df)

# Concatenar todos os DataFrames em um único DataFrame
df_concatenado = pd.concat(dataframes, ignore_index=True)
```

```python
input_path = r'D:\Github\Forked\crypto_technical_analysis\files\crypto_data_with_indicators.parquet'

dados = pd.read_parquet(input_path)

dados['Date'] = pd.to_datetime(dados['Date'])
dados['Date'] = dados['Date'].dt.strftime('%Y-%m-%d')
```

```python
pd.set_option("display.max_columns", None)

dados.tail(2)
```

```python
dados_filter = dados[['Symbol', 'Date', 'Volume', 'Close', 'target_10_7d', 'target_10_15d', 'target_10_30d']]
dados_score = df_concatenado[['Symbol', 'Date', 'lr_pb_10_7d', 'lr_pb_15_7d', 'lr_pb_20_7d', 'lr_pb_25_7d', 'lr_pb_10_15d', 'lr_pb_15_15d', 'lr_pb_20_15d', 'lr_pb_25_15d', 'lr_pb_10_30d', 'lr_pb_15_30d', 'lr_pb_20_30d', 'lr_pb_25_30d', 'score']]
```

```python
import numpy as np

df_merged = pd.merge(dados_filter, dados_score, on=['Date', 'Symbol'], how='inner')

# Substituindo valores infinitos por NaN
df_merged.replace([np.inf, -np.inf], np.nan, inplace=True)

# Removendo linhas com valores NaN
df_merged.dropna(inplace=True)
```

```python
pd.set_option("display.max_rows", 20) 
```

```python
df_merged.sort_values(by='target_10_30d') # retirar o SUI-USD
```

```python
#removendo outliers
df_merged_score_clean = df_merged[df_merged['Symbol'] != 'MYRIA-USD'] ## dados inválidos
df_merged_score_clean.sort_values(by='target_10_15d') # retirar o SUI-USD
df_merged_score_clean.sort_values(by='target_10_15d') # retirar o SUI-USD
df_merged_score_clean.sort_values(by='target_10_30d') # retirar o MON-USD

df_merged_score_clean = df_merged_score_clean[df_merged_score_clean['Symbol'] != 'MON-USD']
df_merged_score_clean = df_merged_score_clean[df_merged_score_clean['Symbol'] != 'SUI-USD']
df_merged_score_clean = df_merged_score_clean[df_merged_score_clean['Symbol'] != 'TIA-USD']
df_merged_score_clean = df_merged_score_clean[df_merged_score_clean['Symbol'] != 'ICE-USD']


df_merged_score_clean
```

```python
# df_merged_score_clean_cumulative = df_merged_score_clean[df_merged_score_clean['score'] > 0.50]
df_merged_score_clean_cumulative = df_merged_score_clean[df_merged_score_clean['Volume'] >= 1000]

df_merged_score_clean_cumulative = df_merged_score_clean_cumulative.sort_values(by='Date') 

df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['lr_pb_25_30d'] > 0.8]
# df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['lr_pb_15_15d'] > 0.7]
# df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['lr_pb_20_7d'] > 0.6]


df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['target_10_7d'] >= -0.95]
df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['target_10_15d'] >= -0.95]
df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['target_10_30d'] >= -0.95]

df_merged_score_clean_cumulative['Close'] = df_merged_score_clean_cumulative['Close'].apply(lambda x: '{:.10f}'.format(x))

# df_merged_score_clean_cumulative = df_merged_score_clean_cumulative[df_merged_score_clean['score'] > 0.4]

df_merged_score_clean_cumulative['Cumulative_Return_7d'] = (1 + ( df_merged_score_clean_cumulative['target_10_7d'])).cumprod() - 1
df_merged_score_clean_cumulative['Cumulative_Return_15d'] = (1 + ( df_merged_score_clean_cumulative['target_10_15d'])).cumprod() - 1
df_merged_score_clean_cumulative['Cumulative_Return_30d'] = (1 + ( df_merged_score_clean_cumulative['target_10_30d'])).cumprod() - 1
df_merged_score_clean_cumulative
```
